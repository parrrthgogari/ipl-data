```python
#dividing the match into powerplay, midlle, and death overs
def over_div(over):
    if over<7:
        return 'Powerplay'
    elif over < 16:
        return 'Middle'
    else:
        return 'Death'

df['phase'] = df['over'].apply(over_div)
df.head(5)
```

```python
df['total_runs'] = df['batsman_runs'] + df['extras']

df['phase'] = df['over'].apply(over_div)

df[['batting_team','bowling_team','batsman','bowler','over','ball','phase','batsman_runs','total_runs']].head(10)
```

```python
#for powerplay
pp = df[df['over'].between(1,6)]

pp['balls_faced'] = 1

batsman_pp = pp.groupby('batsman').agg({
    'batsman_runs': 'sum',
    'balls_faced': 'sum'
}).reset_index()

# Strike Rate = (Runs / Balls) * 100
batsman_pp['SR'] = (batsman_pp['batsman_runs'] / batsman_pp['balls_faced']) * 100

# Keep only batsmen who faced at least 50 balls in Powerplay
batsman_pp = batsman_pp[batsman_pp['balls_faced'] >= 50]

# Sort by Strike Rate
batsman_pp = batsman_pp.sort_values('SR', ascending=False)

batsman_pp.head(10)
```

```python
top10_pp = batsman_pp.head(10)

plt.figure(figsize=(10,6))
plt.bar(top10_pp['batsman'], top10_pp['SR'], color='cyan', edgecolor='black')

plt.title("Best Strike Rate")
plt.xlabel("Batter")
plt.ylabel("Strike Rate")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
death = df[df['over'].between(16,20)]

death['balls_faced'] = 1

batsman_death = death.groupby('batsman').agg({
    'batsman_runs': 'sum',
    'balls_faced': 'sum'
}).reset_index()

batsman_death['SR'] = (batsman_death['batsman_runs'] / batsman_death['balls_faced']) * 100
batsman_death = batsman_death[batsman_death['balls_faced'] >= 30]  # filter small samples

batsman_death = batsman_death.sort_values('batsman_runs', ascending=False)

batsman_death.head(10)
```

```python
batsman_death = batsman_death.sort_values('batsman_runs', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=batsman_death, x='batsman_runs', y='batsman', palette='Reds_r')
plt.title("Top 10 Batsmen in Death Overs (16–20) by Runs", fontsize=14)
plt.xlabel("Runs Scored in Death Overs")
plt.ylabel("Batsman")
plt.show()
```

```python
df['isWide'] = df['isWide'].fillna(0)
balls_faced = df[df['isWide'] == 0]

most_balls_faced = balls_faced.groupby('batsman').size().sort_values(ascending=False).head(10).reset_index(name='balls_faced')
print("--- Top 10 Batsmen by Balls Faced ---")
print(most_balls_faced)

plt.figure(figsize=(12, 7))
sns.barplot(x='balls_faced', y='batsman', data=most_balls_faced, palette='plasma', hue='batsman', legend=False)
plt.title('Top 10 Batsmen by Balls Faced in IPL', fontsize=16)
plt.xlabel('Total Balls Faced', fontsize=12)
plt.ylabel('Batsman', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
runs_conceded = df.groupby('bowler')['total_runs'].sum().reset_index()

valid_wicket_kinds = ['bowled', 'caught', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
wickets_taken = df[df['dismissal_kind'].isin(valid_wicket_kinds)].groupby('bowler').size().reset_index(name='wickets')

bowler_stats = pd.merge(runs_conceded, wickets_taken, on='bowler')

plt.figure(figsize=(10, 6))
sns.scatterplot(data=bowler_stats, x='total_runs', y='wickets')

plt.title('Total Runs Conceded vs. Wickets Taken')
plt.xlabel('Total Runs Conceded')
plt.ylabel('Total Wickets Taken')
plt.grid(True)
plt.show()
```

```python
dot_balls = df[df['total_runs'] == 0]

top_dot_ball_bowlers = dot_balls.groupby('bowler').size().sort_values(ascending=False).head(10).reset_index(name='dot_balls')

plt.figure(figsize=(12, 7))
sns.barplot(x='dot_balls', y='bowler', data=top_dot_ball_bowlers, palette='Blues_r', hue='bowler', legend=False)

plt.title('Top 10 Bowlers by Dot Balls Bowled')
plt.xlabel('Total Dot Balls')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('deliveries_updated_mens_ipl_upto_2024 - deliveries_updated_mens_ipl_upto_2024.csv')

team_name_updates = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Gujarat Titans' : 'Gujarat Lions',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiants': 'Pune',
    'Pune Warriors': 'Pune',
    'Rising Pune Supergiant': 'Pune',
}
df['batting_team'] = df['batting_team'].replace(team_name_updates)
df['bowling_team'] = df['bowling_team'].replace(team_name_updates)

df['total_runs'] = df['batsman_runs'] + df['extras']

team_vs_team_runs = df.groupby(['batting_team', 'bowling_team'])['total_runs'].sum().unstack().fillna(0)

plt.figure(figsize=(12, 10))
sns.heatmap(team_vs_team_runs, annot=True, fmt=".0f", cmap="viridis")

plt.title('Team vs. Team - Total Runs Scored')
plt.xlabel('Bowling Team')
plt.ylabel('Batting Team')
plt.show()
```
